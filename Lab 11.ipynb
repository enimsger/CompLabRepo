```python
import numpy as np
from numpy.random import random as rng
import matplotlib.pyplot as plt

numSteps= 1000

xSteps = 2*(rng(numSteps) > 0.5) - 1
ySteps = 2*(rng(numSteps) > 0.5) - 1
xPosition = np.cumsum(xSteps)
yPosition = np.cumsum(ySteps)

plt.figure()
plt.plot(xPosition, yPosition)
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.axis('Equal')
plt.title('Random Walk')

plt.show()

##-----------------------------------------------------------------
x = 1
plt.figure()
for i in range(4):
    numSteps= 1000

    xSteps = 2*(rng(numSteps) > 0.5) - 1
    ySteps = 2*(rng(numSteps) > 0.5) - 1
    xPosition = np.cumsum(xSteps)
    yPosition = np.cumsum(ySteps)

    plt.subplot(2,2,x)
    plt.plot(xPosition, yPosition)
    plt.xlabel('X Position')
    plt.ylabel('Y Position')
    plt.axis('Equal')
    plt.title('Random Walk')

    x += 1
plt.tight_layout()
plt.show()
##-----------------------------------------------------------------
xFinal = np.zeros(1000)
yFinal = np.zeros(1000)
displacement = np.zeros(1000)
plt.figure()
for i in range(1000):
    numSteps= 1000

    xSteps = 2*(rng(numSteps) > 0.5) - 1
    ySteps = 2*(rng(numSteps) > 0.5) - 1
    xPosition = np.cumsum(xSteps)
    yPosition = np.cumsum(ySteps)
    xFinal[i] = xPosition[-1]
    yFinal[i] = yPosition[-1]
    displacement[i] = np.sqrt(xPosition[-1]**2 + yPosition[-1]**2)
    plt.scatter(xPosition[-1],yPosition[-1])
    
plt.show()
plt.figure()
plt.hist(displacement)
plt.tight_layout()
plt.show()

plt.show()
plt.figure()
plt.hist(displacement**2)
plt.tight_layout()
plt.show()

mean = np.mean(displacement[0]**2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
from scipy.stats import poisson

l_vals = np.arange(0, 50, dtype=float)
rho = (8**l_vals / factorial(l_vals)) * np.exp(-8)
plt.plot(l_vals, rho)
plt.title("A. Plot of (l) = (8^l / l!) * exp(-8)")
plt.xlabel("l")
plt.ylabel("(l)")
plt.grid(True)
plt.show()

N = 1000
flips_per_trial = 100
prob_head = 0.08
trials = np.random.binomial(flips_per_trial, prob_head, size=N)
plt.hist(trials, bins=np.arange(0, max(trials)+2)-0.5, density=True, alpha=0.7, label='Empirical')
plt.title("C. Histogram of heads per trial (N = 1000)")
plt.xlabel("Number of heads (M)")
plt.ylabel("Frequency")
x = np.arange(0, max(trials)+1)
pois_vals = poisson.pmf(x, mu=flips_per_trial * prob_head)
plt.plot(x, pois_vals, 'r-', label='Poisson model')
plt.legend()
plt.grid(True)
plt.show()

N_large = 100000
trials_large = np.random.binomial(flips_per_trial, prob_head, size=N_large)
plt.hist(trials_large, bins=np.arange(0, max(trials_large)+2)-0.5, density=True, alpha=0.7, label='Empirical (large N)')
plt.title("C & D. Histogram vs Poisson (N = 100,000)")
plt.xlabel("Number of heads (M)")
plt.ylabel("Frequency")
x_large = np.arange(0, max(trials_large)+1)
pois_large = poisson.pmf(x_large, mu=flips_per_trial * prob_head)
plt.plot(x_large, pois_large, 'r-', label='Poisson model')
plt.legend()
plt.grid(True)
plt.show()
```
