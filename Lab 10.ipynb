```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt

result_a, _ = quad(lambda x: x**2, 0, 2)
print("a. ∫₀² x² dx =", result_a)

x_vals = np.linspace(0, 5, 100)
integrated_vals = [quad(lambda x: np.exp(-x**2 / 2), 0, x)[0] for x in x_vals]
plt.plot(x_vals, integrated_vals)
plt.title("b. ∫₀ˣ e^(-x²/2) dx from 0 to 5")
plt.xlabel("x")
plt.ylabel("Integral Value")
plt.grid(True)
plt.show()

result_c, _ = quad(lambda x: np.exp(-x**2 / 2), -np.inf, np.inf)
exact = np.sqrt(2 * np.pi)
print("c. ∫₋∞⁺∞ e^(-x²/2) dx =", result_c)
print("Exact value: √(2π) =", exact)
```
